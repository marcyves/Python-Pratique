# Pour quoi Anaconda ?
## parce qu'on peut faire de beaux documents

```python
# Configurer matplotlib pour que les figures s'affichent directement dns ce document
%matplotlib inline

from pylab import *
import numpy as np
```

## Exemple
Une figure simple

```python
x = np.linspace(0, 5, 10)
y = x ** 2

figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('Figure de démo')
show();
```

On peut rendre cela plus lisible avec des couleurs, des symboles.

```python
subplot(1, 2, 1)
plot(x, y, 'r--')

subplot(1, 2, 2)
plot(y, x, 'g*-');
```

On peut faire la même chose avec des API orienté objet

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Un autre magnifique graphique');
```

On peut avoir un controle poussé des axes et des échelles.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Le grand dessin')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Le petit schéma');
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('2 courbes dans le même graphique');
```

```python
# Update the matplotlib configuration parameters:
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('2 courbes et police de caractères améliorée');
```
